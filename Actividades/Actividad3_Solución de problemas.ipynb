```python
# Librerías a utilizar
import matplotlib.pyplot as plt # generar gráficos
import numpy as np # generar arreglos
import pandas as pd # cargar conjuntos de datos externos
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# Cargar conjunto de datos
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100,1) * 2
```

```python
#Preprocesamiento de datos

# Creacion de regresion lineal
train_data, test_data, train_labels, test_labels = train_test_split(
 horas_estudio, calificaciones, test_size=0.33, random_state=50)
```

```python
# Seleccionar el modelo a entrenar: Regresión lineal
model = LinearRegression()
```

```python
# Entrenamiento de modelo de Regresión Lineal
model.fit(horas_estudio, calificaciones)
```

```python
# Realizacion de predicciones
predictions = model.predict(test_data)
```

```python
#Calcular error

mse = mean_squared_error(test_labels, predictions)
print(f"Error Cudratico Medio: {mse}")
```

```python
# Gráfica final con el modelo entrenado

plt.scatter(test_data, test_labels, color='blue',
            label='Datos de Prueba')
plt.plot(test_data, predictions, color='red',
         label='Predicciones del Modelo')

# Area de etiquetación
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Modelo de Regresión Lineal')
plt.legend()
plt.show()
```
